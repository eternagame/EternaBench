```python
# %load /Users/hwayment/ipynb_defaults.py
%load_ext autoreload
%autoreload 2

%pylab inline
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_style('ticks')
sns.set_context('paper')

from matplotlib.colors import LogNorm

from glob import glob
import os, sys, pickle, requests

sys.path.append('../scripts')
import RDAT_utils as utils
import plot_utils as eb_plot
```

```python
outdir = '../assets/CloudLabRound69'

# load in data
data = pd.read_json('../data/full_datasets/Round69_temperaturescan_29Mar2020.json.zip')

#get data corresponding to Round 69

# write dataframe of all sequences concatenated together
concat_data = utils.write_concatenated_dataframe(data)
concat_data = concat_data.loc[concat_data['in_polyA']==0]
# filter outliers
concat_data = utils.filter_data(concat_data)
```

```python
example_package_list = ['vienna_2','nupack_99','rnastructure','contrafold_2','rnasoft_blstar', 'eternafold_B']
titles=['Vienna 2', 'NUPACK 1999', 'RNAstructure', 'CONTRAfold 2', 'RNAsoft BLstar','EternaFold']
```

## Visualize datasets in context of sequences

```python
# all constructs:

figure(figsize=(20,4))
eb_plot.plot_reac_heatplot(data)
savefig('%s/reactivity_heatplot.pdf' % outdir, bbox_inches='tight',transparent=True)
```

```python
# See subset of constructs, also see predicted p(unp) values

figure(figsize=(10,4))
subplot(2,4,1)

eb_plot.plot_reac_heatplot(data, ind_range=[0,60])
colorbar(fraction=0.04)
title('Reactivity')

for i, package in enumerate(example_package_list):
    subplot(2,4,i+2)
    title("p(unp), %s" % titles[i])
    eb_plot.plot_estimate_heatplot(data, ind_range=[0,60], package=package)
    colorbar(fraction=0.04)

tight_layout()
savefig('%s/compare_predicted_heatplots.pdf' % outdir, bbox_inches='tight',transparent=True)
```

```python
# See subset of constructs, also see predicted p(unp) values

figure(figsize=(6,4))

for i, package in enumerate(example_package_list):
    subplot(2,3,i+1)
    title("p(unp), %s" % titles[i])
    scatter(concat_data['p_%s' % package], concat_data['reactivity'], alpha=0.01,color='k',s=20)
    xlabel('predicted p(unp)')
    ylabel('Reactivity')
    #colorbar(fraction=0.04)

tight_layout()
savefig('%s/compare_scatterplots_reac_p.pdf' % outdir, bbox_inches='tight',transparent=True)
```

## Compute correlation between data and package predictions

```python
correlation_data = utils.get_correlations(concat_data, n_bootstraps=20)

# Look at correlations by nucleotides
correlation_data_nucl = utils.get_correlations(concat_data, split_by_nucleotides=True, n_bootstraps=20)
```

```python
correlation_data['package_type'] = correlation_data.apply(lambda row: row['package'].split('_')[0], axis=1)
correlation_data['temperature'] = correlation_data.apply(lambda row: row['package'].split('_')[-1][:-1], axis=1)
```

```python
figure(figsize=(3,3))
tmp_pal=sns.color_palette('Set1',5)
standardized_palette = [tmp_pal[x] for x in [1,2,0,4,3]] + [[218/255,165/255,32/255]] # last is gold for eternafold


sns.pointplot(x='temperature', y='Corr', data=correlation_data,
                      hue='package_type', palette=standardized_palette,
                      join=False, ci='sd', marker='.', scale=0.5, dodge=False)
xticks(rotation=45,horizontalalignment='right')


axhline(0.7360404346538927, color=standardized_palette[-1],label='EternaFold')
axhline(0.7179765071297487, color=standardized_palette[-3],label='CONTRAfold 2')
axhline(0.7242348781651908, color=standardized_palette[-2], label='RNAsoft BLstar')

ylabel('Correlation')
xlabel('Temperature (˚C)')
legend(frameon=False)
savefig('correlation_across_temp.pdf', transparent=True, bbox_inches='tight')
```
