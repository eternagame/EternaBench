## Compute correlation between data and package predictions

```python
correlation_data = utils.get_correlations(concat_data, n_bootstraps=20)

# Look at correlations by nucleotides
correlation_data_nucl = utils.get_correlations(concat_data, split_by_nucleotides=True, n_bootstraps=20)
```

```python
correlation_data = correlation_data.append({'package': 'SPOT-RNA', 'Corr': 0.002 },ignore_index=True)
correlation_data['package_type'] = [x.split('_')[0] for x in correlation_data['package']]

standardized_palette = [tmp_pal[x] for x in [1,2,0,4,3]] + [[0,0,0]]+4*[[128/255, 128/255,128/255]]

figure(figsize=(8,3))
sns.pointplot(x='package', y='Corr', data=correlation_data, order = packages+['SPOT-RNA'],
                      hue='package_type', palette=standardized_palette,
                      join=False, ci='sd', marker='.', scale=0.5, dodge=False)
legend([],frameon=False)
ylabel('Correlation')
xlabel('')
xticks([x for x in range(len(packages)+1)], package_titles+['SPOT-RNA'], rotation=45,horizontalalignment='right')
grid()
savefig('%s/correlation_scatterplot.pdf' % outdir, bbox_inches='tight',transparent=True)
```

```python
figure(figsize=(12,3))
eb_plot.plot_EB(correlation_data_nucl, split_by_nucleotides=True)
legend(bbox_to_anchor=(1,1),frameon=False)
ylabel('Correlation')
savefig('%s/correlation_scatterplot_by_nucl1.pdf' % outdir, bbox_inches='tight',transparent=True)
```

```python
for key in correlation_data['package'].unique():
    avg=np.mean(correlation_data.loc[correlation_data['package']==key]['Corr'])
    std=np.std(correlation_data.loc[correlation_data['package']==key]['Corr'])
    print("%s, %.3f:%.3f" % (key, avg, std))
```
